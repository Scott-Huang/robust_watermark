```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from watermarker import Example_Watermarker
from metric import PSNR, SSIM, Fast_Marching, QMetric
from util import read_img
```

```python
img = read_img('sample_imgs/s1.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
watermark = Example_Watermarker.create_watermark('Hello World', (150, 600))
watermarked_img = Example_Watermarker().watermark(img, (100, 100), watermark)
plt.imshow(watermarked_img)
plt.axis('off')
plt.show()
```

```python
watermark_mask = np.zeros(img.shape[:2], dtype=np.uint8)
watermark_mask[100:100+watermark.shape[0], 100:100+watermark.shape[1]] = 1
qmetric = SSIM()
print(qmetric(watermarked_img, img))
rmetric = Fast_Marching(None, qmetric)
print(rmetric(watermarked_img, watermark_mask, img))
```
